```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

import numpy as np

import warnings

from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score,roc_curve, auc, precision_score, recall_score, f1_score
from sklearn.exceptions import ConvergenceWarning
from sklearn.impute import SimpleImputer

#excell sheet importing
file = r"/home/tamara/Desktop/Heart_Disease_Prediction.csv"
data = pd.read_csv(file)

#handling missing values using SimpleImputer with median
missing_values = data.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)
imputer = SimpleImputer(strategy='median')
data_imputed = data.copy()
numerical_features = data_imputed.select_dtypes(include=['int64', 'float64']).columns
data_imputed[numerical_features] = imputer.fit_transform(data_imputed[numerical_features])

#ensuring categorical variables are properly encoded
if 'Heart Disease' in data_imputed.columns:
    data_imputed['Heart Disease'] = data_imputed['Heart Disease'].apply(lambda x: 1 if x == "Presence" else 0)
if 'Sex' in data_imputed.columns and data_imputed['Sex'].dtype == object:
    encoder = LabelEncoder()
    data_imputed['Sex'] = encoder.fit_transform(data_imputed['Sex'])

#visualizing each numerical attribute and its relationship with heart disease
for feature in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=data_imputed, x=feature, hue='Heart Disease', kde=True, bins=20, palette='coolwarm')
    plt.title(f"Distribution of {feature} by Heart Disease")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

#visualizing categorical features
categorical_features = ['Sex', 'Chest pain type']
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.barplot(x=feature, y='Heart Disease', data=data_imputed, palette="viridis", errorbar=None)
    plt.title(f"Relationship between {feature} and Heart Disease")
    plt.xlabel(feature)
    plt.ylabel("Average Heart Disease (0=Absent, 1=Present)")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


#correlation heatmap
plt.figure(figsize=(10, 8))
corr_matrix = data_imputed.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

#splitting data into features and target
y = data_imputed['Heart Disease']
X = data_imputed.drop('Heart Disease', axis=1)

#stratified K-Fold Cross Validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scaler = StandardScaler()
scaler_ann = MinMaxScaler()

X_train_full, X_test_final, y_train_full, y_test_final = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)


#hyperparameter tuning
param_grid = {
    "Logistic Regression": {"C": [0.1, 1, 10]},
    "Decision Tree": {"max_depth": [3, 5, 10], "min_samples_split": [2, 5, 10]},
    "SVM": {"kernel": ['linear', 'rbf'], "C": [0.1, 1, 10]},
    "KNN": {"n_neighbors": [3, 5, 7, 9]} 
}

def tune_model(model, params, X_train, y_train, metric='accuracy'):
    grid_search = GridSearchCV(model, params, cv=5, scoring=metric, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_


# Initialize and tune models
models = {
    "Logistic Regression": tune_model(LogisticRegression(max_iter=1000, solver='saga'), param_grid["Logistic Regression"], X_train_full, y_train_full, 'f1'),
    "Decision Tree": tune_model(DecisionTreeClassifier(), param_grid["Decision Tree"], X_train_full, y_train_full, 'f1'),
    "SVM": tune_model(SVC(probability=True), param_grid["SVM"], X_train_full, y_train_full, 'f1'),
    "KNN": tune_model(KNeighborsClassifier(), param_grid["KNN"], X_train_full, y_train_full, 'f1')
}

attempt = 0
best_model = None
best_accuracy = 0

while best_accuracy < 0.75 and attempt < 5:

    attempt += 1

    #performance tracking
    model_accuracies = {}
    roc_auc_scores = {}
    roc_curves = {}
    precision_scores = {}
    recall_scores = {}
    f1_scores = {}

    for model_name, model in models.items():
        if model_name == "ANN":
            model_accuracies[model_name] = ann_accuracy
            y_pred_best = (model.predict(X_test_scaled_ann).flatten() > 0.5).astype(int)
            y_pred_proba = model.predict(X_test_scaled_ann).flatten()
        else:
            X_scaled = scaler.fit_transform(X_test_final)
            y_pred = model.predict(X_scaled)
            model_accuracies[model_name] = accuracy_score(y_test_final, y_pred)
            y_pred_best = y_pred

            # Ensure probability predictions for models that support it
            if hasattr(model, "predict_proba"):
                y_pred_proba = model.predict_proba(X_scaled)[:, 1]
            else:
                y_pred_proba = model.decision_function(X_scaled)  # For SVM

        # Compute performance metrics
        precision_scores[model_name] = precision_score(y_test_final, y_pred_best)
        recall_scores[model_name] = recall_score(y_test_final, y_pred_best)
        f1_scores[model_name] = f1_score(y_test_final, y_pred_best)

        # Store ROC curve data
        fpr, tpr, _ = roc_curve(y_test_final, y_pred_proba)
        roc_auc = roc_auc_score(y_test_final, y_pred_proba)

            
        # Compute ROC Curve
        fpr, tpr, _ = roc_curve(y_test_final, y_pred_proba)
        roc_curves[model_name] = (fpr, tpr, roc_auc_score(y_test_final, y_pred_proba))

        accuracies = []
        roc_aucs = []
        precisions = []
        recalls = []
        f1s = []
        mean_fpr = np.linspace(0, 1, 100)
        tprs = []
        
        for train_index, test_index in kf.split(X, y):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]
            
            #scaling the data
            X_train_scaled = scaler.fit_transform(X_train)
            X_test_scaled = scaler.transform(X_test)
            
            #hyperparameter tuning
            if model_name in param_grid:
                model = tune_model(model, param_grid[model_name], X_train_scaled, y_train, metric='f1')

            model.fit(X_train_scaled, y_train)
            y_pred = model.predict(X_test_scaled)
            accuracies.append(accuracy_score(y_test, y_pred))
            precisions.append(precision_score(y_test, y_pred))
            recalls.append(recall_score(y_test, y_pred))
            f1s.append(f1_score(y_test, y_pred))

            
            
            if hasattr(model, "predict_proba"):
                y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
                fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
                tprs.append(np.interp(mean_fpr, fpr, tpr))
                roc_aucs.append(roc_auc_score(y_test, y_pred_proba))
            
        model_accuracies = {}
        for model_name, model in models.items():
           if model_name == "ANN":
            model_accuracies[model_name] = ann_accuracy
           else:
            X_scaled = scaler.fit_transform(X_test_final)
            y_pred = model.predict(X_scaled)
            model_accuracies[model_name] = accuracy_score(y_test_final, y_pred)

        if tprs:
            mean_tpr = np.mean(tprs, axis=0)
            mean_auc = np.mean(roc_aucs)
            roc_curves[model_name] = (mean_fpr, mean_tpr, mean_auc)
    
   # ANN Model
X_train_scaled_ann = scaler_ann.fit_transform(X_train_full)
X_test_scaled_ann = scaler_ann.transform(X_test_final)
ann_model = keras.Sequential([
    layers.InputLayer(input_shape=(X.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X_train_scaled_ann, y_train_full, epochs=50, batch_size=16, verbose=1, validation_data=(X_test_scaled_ann, y_test_final))
models["ANN"] = ann_model
ann_loss, ann_accuracy = ann_model.evaluate(X_test_scaled_ann, y_test_final, verbose=1)


 #finding the best model based on acc
best_model = max(model_accuracies, key=lambda k: model_accuracies[k])
best_accuracy = model_accuracies[best_model]

if best_model in precision_scores:
    print(f"Precision: {precision_scores[best_model]:.4f}")
    print(f"Recall: {recall_scores[best_model]:.4f}")
    print(f"F1-Score: {f1_scores[best_model]:.4f}")
else:
    print(f"Precision, Recall, and F1-Score are not available for {best_model}.")

#plotting roc
plt.figure(figsize=(8, 6))
for model_name, (fpr, tpr, auc_score) in roc_curves.items():
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {auc_score:.2f})")

# Reference line for random guessing
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", alpha=0.7)

plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curves for Different Models")
plt.legend()
plt.grid(True)
plt.show()


#filtering out the models that lack metric values (e.g., ANN might not have them)
valid_models = [m for m in model_accuracies.keys() if m in precision_scores]

print("Model Accuracies:", model_accuracies)
print("Precision Scores:", precision_scores)
print("Recall Scores:", recall_scores)
print("F1 Scores:", f1_scores)

model_performance_df = pd.DataFrame({
    "Model": list(model_accuracies.keys()),
    "Accuracy": list(model_accuracies.values()),
    "Precision": list(precision_scores.values()),
    "Recall": list(recall_scores.values()),
    "F1-Score": list(f1_scores.values()),
}).sort_values(by="Accuracy", ascending=False)

print("\nModel Performance Summary:")
print(model_performance_df)


#plotting model comparison
plt.figure(figsize=(10, 6))
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
colors = ['blue', 'green', 'red', 'purple']

#confusion matrix for the best model
print(f"\nConfusion Matrix for {best_model}:")

#ensure correct final test data
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_test_scaled_final = scaler.transform(X_test_final)
if best_model == "ANN":
    X_test_scaled_final = scaler_ann.transform(X_test_final)

#retrieving the trained model
if best_model == "ANN":
    best_trained_model = ann_model
    cm = confusion_matrix(y_test_final, (best_trained_model.predict(X_test_scaled_final).flatten() > 0.5).astype(int))

else:
    best_trained_model = tune_model(models[best_model], param_grid[best_model], X_train_final, y_train_final, metric='f1')
    best_trained_model.fit(X_train_final, y_train_final)  # Ensure the model is trained
    y_pred_best = best_trained_model.predict(X_test_scaled_final)

#confusion matrix for the best model
print(f"\nConfusion Matrix for {best_model}:")
cm = confusion_matrix(y_test_final, (best_trained_model.predict(X_test_scaled_final).flatten() > 0.5).astype(int))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Disease", "Disease"])

plt.figure(figsize=(6, 5))
disp.plot(cmap="Blues", values_format='d')
plt.title(f"Confusion Matrix for {best_model}")
plt.show()


for i, metric in enumerate(metrics):
    plt.barh(model_performance_df["Model"], model_performance_df[metric], color=colors[i], alpha=0.7, label=metric)

plt.xlabel("Score")
plt.title("Model Performance Comparison")
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

#featuring importance for decision tree and logistic regression

if best_model in ["Decision Tree", "Logistic Regression"]:
    best_trained_model = models[best_model]
    if hasattr(best_trained_model, "feature_importances_"):
        importances = best_trained_model.feature_importances_
    else:
        importances = np.abs(best_trained_model.coef_).flatten()
    
    feature_importance_df = pd.DataFrame({"Feature": X.columns, "Importance": importances})
    feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Importance", y="Feature", data=feature_importance_df, palette="viridis")
    plt.title(f"Feature Importance for {best_model}")
    plt.show()



#function to allow user input and prediction
def predict_heart_disease(model, scaler, model_name):
    print(f"\n--- Predict Heart Disease with {model_name} ---")
    try:
        user_input = [[
            float(input("Enter Age: ")), int(input("Enter Sex (1 for Male, 0 for Female): ")),
            int(input("Enter Chest Pain Type (1-4, where 4 is the most severe): ")),
            float(input("Enter Resting Blood Pressure: ")), float(input("Enter Cholesterol Level: ")),
            int(input("Enter Fasting Blood Sugar > 120 mg/dl (1 = Yes, 0 = No): ")),
            int(input("Enter Resting Electrocardiographic Results (0, 1, 2): ")),
            float(input("Enter Maximum Heart Rate Achieved: ")), int(input("Enter Exercise Induced Angina (1 = Yes, 0 = No): ")),
            float(input("Enter ST Depression Induced by Exercise: ")), int(input("Enter Slope of the Peak Exercise ST Segment (1-3): ")),
            int(input("Enter Number of Major Vessels (0-3): ")), int(input("Enter Thalassemia (1 = Normal, 2 = Fixed Defect, 3 = Reversible Defect): "))
        ]]
        user_input_scaled = scaler.transform(user_input)
        prediction = model.predict(user_input_scaled)
        print(f"Prediction: {'Likely has heart disease' if prediction[0] else 'Unlikely to have heart disease'}")
    except Exception as e:
        print(f"Error: {e}")

#calling prediction function for the best model
predict_heart_disease(models[best_model], scaler, best_model) 
```
